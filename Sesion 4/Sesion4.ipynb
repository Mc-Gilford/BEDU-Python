# Series
* Tienen un indice y se visualiza
* Solo se guarda un tipo de dato

```
import pandas as pd

serie_a = pd.Series([2,3,4,6,5])
serie_a
```

```
serie_b = pd.Series( [1,2,3,4,5], index=[12,13,14,15,16])
serie_b
```

```
serie_c = pd.Series( [1,2,3,4,5], index=['a','b','c','d','f'])
serie_c
```

```
serie_c.loc['d']
```

```
datos = {
    "Juan": 45,
    "Pepe": 56,
    "Alfonsina": 12,
    "Jenny": 49,
    "Marco P.": 12
}

serie_d = pd.Series(datos)
serie_d
```

```
print(serie_a)

serie_a.iloc[[1,3,4]]
```

```
print(serie_c)
serie_c.loc[['a','b','f']]
```

# Modificacion dataframes

```
datos = {
    "nombre": ["Jose", "Angel" ,"Meliodas", "Palmer", "Sara"],
    "cereal": ["Choco", "Zuca", "Korn", "Fruit", "Nestil"]
}
df = pd.DataFrame(datos)
df
```

```
hr_des = [11,12,3,10,12]
df["Hora del desayuno"] = pd.Series(hr_des)
df
```

```
df=df.drop(columns="Hora del desayuno")
df
```

```
df=df.drop(index=0)
df
```

```
dato = {
    "nombre": ["kara"],
    "cereal": ["cruji"]
}

df1 = df.append(pd.DataFrame(dato))
df1
```
