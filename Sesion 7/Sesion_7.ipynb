# Apply

```
tacos = {
    "Sergio":[4,10],
    "Monica":[5,8],
    "Jose":[6,8],
    "Zoe":[5,12],
    "Jesus":[4,22],
    "Emilio":[6,14]
}

df_tacos = pd.DataFrame.from_dict(tacos)

df_tacos
```

```
def promedio(val):
  return sum(val)/len(val)

df_tacos.apply(promedio, axis=1)
```

```
df_tacos.apply(promedio, axis=0)
```

```
filtro_mayor_a_15=bestsellers["price.numberDouble"]>15  
filtro_inicia_con_r=bestsellers["author"].str.startswith("R")
libros=bestsellers[filtro_mayor_a_15 & filtro_inicia_con_r]
libros.head()
```
