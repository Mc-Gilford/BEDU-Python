# Lista


```
Lista_1 = [1,2,3,4]
print(Lista_1)

print(Lista_1[0])
len(Lista_1) #Nos dice el numero de elementos
print(Lista_1[-1]) #Nos extrae los datos de la derecha con el -
```

```
var_1 = 2
var_2 = 4
var_3 = 6
var_4 = "Sergio"
Lista_2 = [var_1,var_2,var_3,var_4]
print(Lista_2)
```

```
lista_con_listas = [[1,2,3], Lista_1, Lista_2]
print(lista_con_listas)
print(lista_con_listas[-2][-1])
```

# Modificando listas


```
Tupla_1 = (1,4,5,1)
print(Tupla_1)
Lista_1.append(6)
print(Lista_1) 
Lista_1.pop()
print(Lista_1)
```

```
#insert y pop
Lista_1.insert(3,6)
print(Lista_1)
Lista_1.pop(3)
print(Lista_1)
Lista_1 [2] = 5
print(Lista_1)
```

```
#Las tuplas no se pueden modificar
#var.split("")
print(Lista_1[3:]) #: nos  sirve para extraer dentro de un parametro
```
