# Limpiando datos

```
tienda = {
    'precio':[100,201,np.nan,400,500],
    'cant_stock':[10,20,40,20,13],
    'product_vendidos':[100,12,12,3,np.nan]
}

df = pd.DataFrame(tienda,index=["Tacos", "Jabon", "Atun", "TV", "Salmon"])
df
```

```
df.isna()
```

```
df.isna().sum(axis=1)
```

```
df.dropna()
```

```
df.fillna(0)
```

```
df["product_vendidos"] = df["product_vendidos"].fillna(0)
df
```

```
df.dropna(axis=0)
```
